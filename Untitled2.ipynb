Read files from Google Drive and create DataFrames


```python
import os
import gdown
from pyspark.sql import SparkSession


# Create a SparkSession
spark = SparkSession.builder.appName("Weather Forecast").getOrCreate()

# Google Drive folder URL
folder_url = "https://drive.google.com/drive/folders/1zHHhaKqmUTvZMA1O8Vvn73CRYjrqSX_S?usp=drive_link"

#
# Fetch file list in the folder
file_list = gdown.download_folder(folder_url, quiet=False)
```

```python
dfs = []

# Process each file and create a DataFrame
for file_path in file_list:
    # Extract file name without extension
    file_name = os.path.splitext(os.path.basename(file_path))[0]

    # Create DataFrame and append it to the list
    df = spark.read.csv(file_path, header=True, inferSchema=True)
    dfs.append((f"df_{file_name}", df))

# Optionally, you can perform operations on the DataFrames
for df_name, df in dfs:
    print(f"DataFrame {df_name}:")
    df.show(5)
```

Change the structure of Attribute tables into the same as City table, join all the tables as one

```python
from pyspark.sql.functions import col, expr
df=dfs[0][1]
for i in range(1, len(dfs)):
    df_name, df = dfs[i]

    # Get the list of city columns dynamically
    city_columns = [col_name for col_name in df.columns if col_name != "datetime"]

    # Build the stack expression dynamically with backticks for column names with spaces
    stack_expr = f"stack({len(city_columns)}, " + ", ".join([f"'{col_name}', `{col_name}`" for col_name in city_columns]) + f") as (city_name, {df_name})"

    # Melt the DataFrame
    melted_df = df.selectExpr("datetime", stack_expr)

    # Replace the original DataFrame with the melted DataFrame in the list
    dfs[i] = (df_name, melted_df)
print(dfs)

```

```python
df = dfs[0][1]
df = df.withColumnRenamed("City", "city_name")
temp=dfs[1][1]
for i in range(2, len(dfs)):
    df_name, melted_df = dfs[i]
    temp = temp.join(melted_df, ["city_name","datetime"], "inner")
df=df.join(temp,["city_name"])
temp.printSchema()
df.printSchema()
print("The count of final data: " + str(df.count()))
```

There are 32 different outcomes so we decided to recategorize it into some main types of weather

```python
from pyspark.sql.functions import when
def categorize_weather(old_category):
    if old_category in ['sky is clear']:
        return 'Clear Sky'
    elif old_category in ['few clouds', 'broken clouds', 'overcast clouds', 'scattered clouds']:
        return 'Cloudy'
    elif old_category in ['very heavy rain', 'proximity shower rain', 'heavy shower rain',
                          'heavy intensity shower rain', 'freezing rain', 'light rain', 'light intensity drizzle',
                          'light intensity shower rain', 'moderate rain', 'heavy intensity rain',
                          'light intensity drizzle rain', 'shower rain', 'rain and snow']:
        return 'Rainy'
    elif old_category in ['heavy shower snow', 'shower snow', 'light snow', 'snow',
                          'light rain and snow', 'light shower snow', 'heavy snow']:
        return 'Snowy'
    elif old_category in ['proximity thunderstorm', 'thunderstorm with rain',
                          'thunderstorm with heavy rain', 'thunderstorm with light rain', 'thunderstorm']:
        return 'Stormy'
    elif old_category in ['fog', 'mist']:
        return 'Foggy'
    elif old_category == 'haze':
        return 'Haze'
    else:
        return old_category

categorize_weather_udf = spark.udf.register("categorize_weather", categorize_weather)
df_toronto = df_toronto.withColumn("weather_category", categorize_weather_udf(df_toronto["df_weather_description"]))

df_toronto.show(5,truncate=False)
description = df_toronto.groupBy("weather_category").agg(count("*").alias("record_count"))
description = description.orderBy("record_count")
print(description.count())
description.show(truncate=False)
```

Break the datetime column into hour, day, month and year columns since we assume weather data is kinda time-series

```python
from pyspark.sql.functions import hour,month,year,day
df_toronto = df_toronto.withColumn("hour", hour("datetime"))
df_toronto = df_toronto.withColumn("day", day("datetime"))
df_toronto = df_toronto.withColumn("month", month("datetime"))
df_toronto = df_toronto.withColumn("year", year("datetime"))
df_toronto.printSchema()
df_toronto.show(5)
```

We split the data into 4 seasons according to Toronto climate to find out if there is any difference between seasons

```python
import matplotlib.pyplot as plt
def categorize_season(month, day):
    if (month == 3 and day >= 20) or (month == 4 or month == 5) or (month == 6 and day <= 21):
        return "Spring"
    elif (month == 6 and day >= 21) or (month == 7 or month == 8) or (month == 9 and day <= 22):
        return "Summer"
    elif (month == 9 and day >= 23) or (month == 10 or month == 11) or (month == 12 and day <= 22):
        return "Fall"
    else:
        return "Winter"

categorize_season_udf = spark.udf.register("categorize_season", categorize_season)
df_toronto = df_toronto.withColumn("season", categorize_season_udf(df_toronto["month"], df_toronto["day"]))
grouped_df = df_toronto.groupBy("year", "season", "weather_category").count()

data_to_plot = grouped_df.collect()
print(data_to_plot)
```

```python
for category in df_toronto.select("weather_category").distinct().collect():
    category = category[0]

    plt.figure(figsize=(6, 3))
    plt.title(f'Total {category} By Season and Year')
    plt.xlabel('Season')
    plt.ylabel('Total Count')

    for year in df_toronto.select("year").distinct().collect():
        year = year[0]
        category_data = [row['count'] for row in data_to_plot if row['weather_category'] == category and row['year'] == year]
        plt.plot(category_data, label=f"{year}")

    plt.xticks(range(4), ['Spring', 'Summer', 'Fall', 'Winter'])
    plt.legend()
    plt.show()
```

Doing the same thing with 'hour columns', Toronto climate shows that there is difference between day and night time in the first half and second half of a year. We label each row as day and night to find out if there is any pattern

```python
daytime_condition = (
    ((col("season") == "Spring") | (col("season") == "Summer")) &
    (col("hour") >= 6) &
    (col("hour") <= 20)
) | (
    ((col("season") == "Fall")|(col("season") == "Winter")  &
    (col("hour") >= 7) &
    (col("hour") <= 18)
))
df_toronto = df_toronto.withColumn("daytime", when(daytime_condition, "Day").otherwise("Night"))
df_toronto.show(5,truncate=False)
```

```python
# Group by year, season, day, and weather category to get the count
# grouped_daytime_df = df_toronto.groupBy("year", "daytime", "weather_category").count()
# grouped_daytime_df.show()
grouped_daytime_df = df_toronto.groupBy("daytime", "weather_category","year").count()
grouped_daytime_df.show()
```

```python
from pyspark.sql import functions as F
import numpy as np

for category in grouped_daytime_df.select("weather_category").distinct().collect():
    category=category[0]
    temp = grouped_daytime_df.filter(grouped_daytime_df.weather_category == category)
    sum_day = temp.filter(temp.daytime == 'Day').groupBy('year').agg(F.sum('count').alias('sum_count_day'))
    sum_night = temp.filter(temp.daytime == 'Night').groupBy('year').agg(F.sum('count').alias('sum_count_night'))
    schema = sum_day.join(sum_night, on='year', how='outer').fillna(0)
    bar_width = 0.35
    bar_positions1 = np.arange(schema.count())
    bar_positions2 = bar_positions1 + bar_width

    fig, ax = plt.subplots()
    ax.bar(bar_positions1, schema.select('sum_count_day').rdd.flatMap(lambda x: x).collect(), width=bar_width, label='Day')
    ax.bar(bar_positions2, schema.select('sum_count_night').rdd.flatMap(lambda x: x).collect(), width=bar_width, label='Night')

    ax.set_xlabel('Year')
    ax.set_ylabel('Count')
    ax.set_title(f'Double Bar Chart for {category} Weather Category')
    ax.set_xticks(bar_positions1 + bar_width / 2)
    ax.set_xticklabels([str(row.year) for row in schema.collect()])
    ax.legend()

    plt.show()
```

We found out some patterns about day and night data so we decided to put colums to note it is at night or in the daytime

```python
df_encoded= df_toronto.withColumn("isday", when(col("daytime") == "Day", 1).otherwise(0))
df_encoded= df_encoded.withColumn("isnight", when(col("daytime") == "Night", 1).otherwise(0))
df_encoded.show(1, truncate=False)
df_encoded.printSchema()
```

```python
from pyspark.ml.feature import OneHotEncoder, StringIndexer, VectorAssembler
from pyspark.ml import Pipeline
from pyspark.ml.classification import RandomForestClassifier
from pyspark.ml.evaluation import MulticlassClassificationEvaluator

target = "weather_category"
label_indexer = StringIndexer(inputCol="weather_category", outputCol="label", handleInvalid="skip")
df_encoded = label_indexer.fit(df_encoded).transform(df_encoded)

if "features" in df_encoded.columns:
    df_encoded = df_encoded.drop("features")

# Create a VectorAssembler to assemble features into a vector column
assembler = VectorAssembler(inputCols=["df_humidity","df_pressure","df_temperature","day","month","df_wind_speed","year","isday","isnight"]
, outputCol="features")
df_encoded = assembler.transform(df_encoded)

# Split the data into training and testing sets
(trainingData, testData) = df_encoded.randomSplit([0.8, 0.2], seed=1234)
```

```python
from pyspark.ml.tuning import ParamGridBuilder, CrossValidator
rf_cross_validator = RandomForestClassifier(labelCol="label", featuresCol="features")
param_grid = ParamGridBuilder().addGrid(rf_cross_validator.numTrees, [10, 20, 30]).addGrid(rf_cross_validator.maxDepth, [5, 10, 15]).build()
cross_validator = CrossValidator(estimator=rf_cross_validator,
                                 estimatorParamMaps=param_grid,
                                 evaluator=MulticlassClassificationEvaluator(labelCol="label", predictionCol="prediction", metricName="accuracy"),
                                 numFolds=5,
                                 seed=1234)
cross_model = cross_validator.fit(trainingData)
cross_predictions = cross_model.transform(testData)

# Evaluate the cross-validated model
cross_evaluator = MulticlassClassificationEvaluator(labelCol="label", predictionCol="prediction", metricName="accuracy")
cross_accuracy = cross_evaluator.evaluate(cross_predictions)
print(f"Cross-Validated Model Accuracy: {cross_accuracy}")

# Calculate F1-score, recall, and precision for the cross-validated model
f1_cross = cross_evaluator.evaluate(cross_predictions, {cross_evaluator.metricName: "f1"})
recall_cross = cross_evaluator.evaluate(cross_predictions, {cross_evaluator.metricName: "weightedRecall"})
precision_cross = cross_evaluator.evaluate(cross_predictions, {cross_evaluator.metricName: "weightedPrecision"})
print(f"F1-score (Cross-Validated Model): {f1_cross}")
print(f"Recall (Cross-Validated Model): {recall_cross}")
print(f"Precision (Cross-Validated Model): {precision_cross}")
```
